```python
import seaborn as sns 
```

```python
tips= sns.load_dataset('tips')
```

```python
tips.head()
```

```python
tips.shape
```

```python
tips.dtypes
```

```python
sns.pairplot(tips,hue='sex',palette='husl')
```

```python
sns.distplot(tips['tip'],color='green')
```

```python
sns.barplot(x='total_bill',y='sex',data=tips,palette='coolwarm')
```

```python
sns.barplot(x="day",y="tip",hue='sex',data=tips, palette="magma")
```

```python
sns.countplot(tips['sex'],palette='muted')
```

```python
tips['sex'].value_counts()
```

```python
sns.countplot('day',data=tips,hue='smoker',palette='hls')
```

```python
sns.violinplot(x='sex',y='total_bill',data=tips,palette= 'magma')
```

```python
sns.violinplot(x='total_bill',y='day',hue='time',palette='magma', data=tips)
```

```python
sns.boxplot('smoker','total_bill',data=tips, palette='husl')
```

```python
sns.jointplot(x='tip',y='total_bill', data=tips, kind='hex')
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='reg')
```

```python
sns.catplot(x='smoker',y="tip",hue="sex",data=tips,kind='swarm')
```
